```python
import numpy as np
import torch
import tifffile
import sys
sys.path.append('/net/mulan/home/huiwann/spatial/SVC/')
from model.utils import *
```

```python
data_path = '/net/mulan/home/huiwann/spatial/SVC/' 
dataset = 'data/merfish_U2OS' 
device = 'cuda:1'

gene_names = np.loadtxt(f'{data_path}{dataset}/gene_names.txt', dtype=str)
cell_names = np.loadtxt(f'{data_path}{dataset}/cell_names.txt', dtype=str)

data = np.load(f'{data_path}{dataset}/cell_gene_map_low_res.npz')["image"]
print("shape of total map of low resolution:",data.shape)

location =  np.load(f"{data_path}{dataset}/cell_location.npy") 

cell_morphology = np.load(f"{data_path}{dataset}/cell_morphology.npy") 
nuclear_morphology = np.load(f"{data_path}{dataset}/nuclear_morphology.npy")  
print(cell_morphology.shape) 
print(nuclear_morphology.shape)

```

```python
batch=('-61','-62','-63','-64','-65','-66','-67','-68')
train_indices,test_indices = [],[]
for i, s in enumerate(cell_names):
    if s.endswith(batch):
        test_indices.append(i)
    else:
        train_indices.append(i)
train_image, train_data_location, train_cell_names, train_cell_morphology, train_nuclear_morphology = data[train_indices,:,:,:], location[train_indices,:], np.array(cell_names)[train_indices], cell_morphology[train_indices], nuclear_morphology[train_indices]
test_image, test_data_location, test_cell_names, test_cell_morphology, test_nuclear_morphology = data[test_indices,:,:,:], location[test_indices,:], np.array(cell_names)[test_indices], cell_morphology[test_indices], nuclear_morphology[test_indices]
print("shape of train image:",train_image.shape)
print("shape of test image:",test_image.shape)
# train_count_sum = train_image.sum(axis=(2,3))
# np.save(f'{data_path}output/merfish_U2OS/train_count_sum.npy', train_count_sum.cpu().numpy())
```

```python
np.savez_compressed(f"{data_path}{dataset}/train_merfish_U2OS.npz", data_ori=train_image, cell_morphology=train_cell_morphology, nuclear_morphology=train_nuclear_morphology, location=train_data_location, cell_names=train_cell_names)
np.savez_compressed(f"{data_path}{dataset}/test_merfish_U2OS.npz", data_ori=test_image, cell_morphology=test_cell_morphology, nuclear_morphology=test_nuclear_morphology, location=test_data_location, cell_names=test_cell_names)
```
